```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
from selenium.common.exceptions import StaleElementReferenceException

# Inicializar el navegador
driver = webdriver.Chrome()

# Estructura de categorías y subcategorías
categorias = {
    'Electro-y-tecnologia': {
                'Lavado': ['Lavarropas', 'Secarropas', 'Lavasecarropas', 'Lavavajillas'],
                'Smart-TV-y-soportes': ['Smart-TV', 'Soportes-y-accesorios'],
    },
}

# Función para navegar por las categorías y subcategorías
def obtener_productos_por_categoria(categoria, subcategoria=None, subsubcategoria=None):
    # Construir la URL según la categoría, subcategoría y subsubcategoría
    if subsubcategoria:
        url = f'https://www.carrefour.com.ar/{categoria}/{subcategoria}/{subsubcategoria}'
    elif subcategoria:
        url = f'https://www.carrefour.com.ar/{categoria}/{subcategoria}'
    else:
        url = f'https://www.carrefour.com.ar/{categoria}'

    # Acceder a la página
    driver.get(url)

    # Buscar los productos
    product_elements = driver.find_elements(By.CSS_SELECTOR, '#slider-items-02rjh1z div')

    # Recopilar datos de los productos
    data = []
    for element in product_elements:
        try:
            titulo_element = element.find_element(By.TAG_NAME, 'h3')
            precio_element = element.find_element(By.CLASS_NAME, 'valtech-carrefourar-product-price-0-x-currencyContainer')
            imagen_element = element.find_element(By.TAG_NAME, 'img').get_attribute('src')

            data.append({
                'Titulo': titulo_element.text,
                'Precio': precio_element.text,
                'Imagen': imagen_element
            })

        except NoSuchElementException:
            pass  # Ignora elementos que no tienen los datos esperados

        except StaleElementReferenceException:
            pass  # Ignora elementos obsoletos

    # Conversion de un DataFrame de pandas
    df = pd.DataFrame(data)

    return df


df = obtener_productos_por_categoria(categoria, subcategoria, subsubcategoria)
print(df)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException, TimeoutException, ElementClickInterceptedException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
import logging

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s: %(message)s')
logger = logging.getLogger(__name__)

# Configuración de selectores y páginas (mismo diccionario que antes)
paginas = {
    'carrefour': {
        'url_base': 'https://www.carrefour.com.ar',
        'selectores': {
            'titulo': '.vtex-product-summary-2-x-productBrand.vtex-product-summary-2-x-brandName.t-body',
            'precio': '.valtech-carrefourar-product-price-0-x-currencyContainer',
            'imagen': '.vtex-product-summary-2-x-imageNormal',
            'boton_siguiente': '.valtech-carrefourar-search-result-2-x-paginationButtonPages button'
        },
        'categorias': {
            'Electro-y-tecnologia': {
                'Lavado': ['Lavarropas', 'Secarropas', 'Lavasecarropas', 'Lavavajillas'],
                'Smart-TV-y-soportes': ['Smart-TV', 'Soportes-y-accesorios'],
            },
        }
    }
    # Puedes agregar más páginas aquí
}

def close_overlays(driver):
    """
    Intenta cerrar elementos emergentes que puedan bloquear la interacción
    """
    overlay_selectors = [
        '.cookie-consent-close',
        '.close-popup',
        '#popup-close',
        '.modal-close'
    ]
    
    for selector in overlay_selectors:
        try:
            overlay = WebDriverWait(driver, 3).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, selector))
            )
            overlay.click()
        except:
            pass

def wait_and_click(driver, locator, timeout=10):
    """
    Método robusto para hacer clic en un elemento
    """
    try:
        element = WebDriverWait(driver, timeout).until(
            EC.element_to_be_clickable(locator)
        )
        
        # Scroll into view
        driver.execute_script("arguments[0].scrollIntoView({block: 'center', behavior: 'smooth'});", element)
        
        # Intentar hacer clic de diferentes maneras
        try:
            element.click()
        except:
            try:
                driver.execute_script("arguments[0].click();", element)
            except Exception as e:
                logger.error(f"No se pudo hacer clic en el elemento: {e}")
                return False
        
        return True
    
    except TimeoutException:
        logger.warning(f"Timeout esperando el elemento: {locator}")
        return False

def obtener_productos_con_paginacion(pagina, categoria, subcategoria=None, subsubcategoria=None):
    """
    Extrae productos con manejo robusto de paginación
    """
    # Inicializar navegador
    options = webdriver.ChromeOptions()
    # options.add_argument('--headless')  # Descomentar para modo sin interfaz
    options.add_argument('--start-maximized')
    options.add_argument('--disable-extensions')
    
    driver = webdriver.Chrome(options=options)
    
    try:
        # Configuración específica
        config = paginas[pagina]
        url_base = config['url_base']
        selectores = config['selectores']
        
        # Construcción de la URL
        if subsubcategoria:
            url = f'{url_base}/{categoria}/{subcategoria}/{subsubcategoria}'
        elif subcategoria:
            url = f'{url_base}/{categoria}/{subcategoria}'
        else:
            url = f'{url_base}/{categoria}'
        
        driver.get(url)
        
        # Esperar carga inicial
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, selectores['titulo']))
        )
        
        # Cerrar posibles overlays
        close_overlays(driver)
        
        data = []
        pagina_actual = 1
        
        # Obtener la cantidad total de páginas
        try:
            botones_paginacion = driver.find_elements(By.CSS_SELECTOR, '.valtech-carrefourar-search-result-2-x-paginationButtonPages button')
            total_paginas = len(botones_paginacion)
            logger.info(f"Total de páginas: {total_paginas}")
        except Exception as e:
            logger.warning(f"No se pudo obtener el número total de páginas: {e}")
            total_paginas = 3  # Valor predeterminado
        
        while pagina_actual <= total_paginas:
            logger.info(f"Extrayendo datos de la página {pagina_actual} para {pagina}...")
            
            try:
                # Extraer productos
                nombres = driver.find_elements(By.CSS_SELECTOR, selectores['titulo'])
                precios = driver.find_elements(By.CSS_SELECTOR, selectores['precio'])
                imagenes = driver.find_elements(By.CSS_SELECTOR, selectores['imagen'])
                
                for i in range(min(len(nombres), len(precios), len(imagenes))):
                    data.append({
                        'Titulo': nombres[i].text.strip(),
                        'Precio': precios[i].text.strip(),
                        'Imagen': imagenes[i].get_attribute('src').strip()
                    })
                
                # Salir si es la última página
                if pagina_actual == total_paginas:
                    break
                
                # Navegar a la siguiente página
                try:
                    siguiente_boton = driver.find_element(By.CSS_SELECTOR, f'button[value="{pagina_actual + 1}"]')
                    
                    # Scroll y clic con estrategias múltiples
                    driver.execute_script("arguments[0].scrollIntoView({block: 'center', behavior: 'smooth'});", siguiente_boton)
                    
                    max_intentos = 3
                    for intento in range(max_intentos):
                        try:
                            WebDriverWait(driver, 10).until(EC.element_to_be_clickable(siguiente_boton)).click()
                            break
                        except:
                            try:
                                driver.execute_script("arguments[0].click();", siguiente_boton)
                                break
                            except Exception as e:
                                logger.warning(f"Intento {intento + 1} fallido: {e}")
                                if intento == max_intentos - 1:
                                    logger.error("No se pudo navegar a la siguiente página")
                                    break
                    
                    # Esperar carga de la página
                    WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, selectores['titulo']))
                    )
                    
                    time.sleep(2)  # Pequeña pausa para estabilidad
                    pagina_actual += 1
                
                except Exception as e:
                    logger.error(f"Error al navegar: {e}")
                    break
            
            except Exception as e:
                logger.error(f"Error en la página {pagina_actual}: {e}")
                break
        
        # Convertir a DataFrame
        df = pd.DataFrame(data)
        return df
    
    except Exception as e:
        logger.error(f"Error general: {e}")
        return pd.DataFrame()
    
    finally:
        driver.quit()

# Ejemplo de uso
def main():
    try:
        pagina = 'carrefour'
        categoria = 'Electro-y-tecnologia'
        subcategoria = 'Lavado'
        subsubcategoria = 'Lavarropas'
        
        df = obtener_productos_con_paginacion(pagina, categoria, subcategoria, subsubcategoria)
        
        if not df.empty:
            print(df)
            # Opcional: guardar en CSV
            df.to_csv(f'productos_{pagina}_{subcategoria}.csv', index=False, encoding='utf-8-sig')
        else:
            print("No se encontraron productos")
    
    except Exception as e:
        logger.error(f"Error en la ejecución principal: {e}")

if __name__ == "__main__":
    main()
```
